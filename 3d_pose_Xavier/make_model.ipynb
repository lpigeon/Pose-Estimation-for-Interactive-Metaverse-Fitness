```python
import pandas as pd
import numpy as np
```

```python
import sklearn
```

```python
data_UP = pd.read_csv('./src/Up_data.csv')
data_Down1 = pd.read_csv('./src/Down_data1.csv')
data_Down2 = pd.read_csv('./src/Down_data2.csv')
data_Down3 = pd.read_csv('./src/Down_data3.csv')
data_Down4 = pd.read_csv('./src/Down_data4.csv')
```

```python
data_Down = pd.concat([data_Down1,data_Down2,data_Down3,data_Down4],axis=0)
```

```python
print(data_UP.shape)
print(data_Down.shape)
```

```python
data_UP.columns
```

```python
data_UP = data_UP.drop(['Unnamed: 0'],axis=1)
data_Down = data_Down.drop(['Unnamed: 0'],axis=1)
```

```python
data_UP['state'] = 'UP'
data_Down['state'] = 'Down'
```

```python
Datas = pd.concat([data_UP,data_Down],axis=0)
Datas.shape
```

```python
Datas
```

```python
Datas = Datas.sample(frac=1).reset_index(drop=True) 
```

```python
Datas
```

```python
y = Datas['state']
x = Datas.drop(['state'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier

logistic_clf = LogisticRegression(random_state=777,C= 10,solver='liblinear')
logistic_clf.fit(X_train, y_train)

KNN_clf = KNeighborsClassifier(n_neighbors=20)
KNN_clf.fit(X_train, y_train)

NB_clf = GaussianNB()
NB_clf.fit(X_train, y_train)

RF_clf = RandomForestClassifier(max_depth=7, random_state=0)
RF_clf.fit(X_train, y_train)
```

```python
a = np.array([X_test.iloc[0]])
```

```python
a
```

```python
logistic_clf.predict(a)
```

```python
logistic_preds= logistic_clf.predict(X_test)
knn_preds = KNN_clf.predict(X_test)
NB_preds = NB_clf.predict(X_test)
rf_preds = RF_clf.predict(X_test)

i = 105
print(logistic_preds[i])
print(knn_preds[i])
print(NB_preds[i])
print(rf_preds[i])

print(y_test.iloc[i])
```

```python
from sklearn.metrics import accuracy_score
print('logistic_ACC.....',accuracy_score(y_test, logistic_preds))
print('knn ACC..........',accuracy_score(y_test, knn_preds))
print('nb_ACC...........',accuracy_score(y_test, NB_preds))
print('rf_ACC...........',accuracy_score(y_test, rf_preds))
```

```python
import joblib

joblib.dump(logistic_clf, 'Logistic.pkl')
joblib.dump(KNN_clf, 'KNN.pkl')
joblib.dump(NB_clf, 'NB.pkl')
joblib.dump(RF_clf, 'RF.pkl')
```
